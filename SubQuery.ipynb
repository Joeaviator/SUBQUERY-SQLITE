```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
database='database.sqlite'
conne=sqlite3.connect(database)
print("Database connected successfully")
```

```python
#Read the tables
tables=pd.read_sql("""SELECT * FROM sqlite_master WHERE type='table'""",conne)
tables
```

```python
team= pd.read_sql("""SELECT * FROM Team""",conne)
team
```

```python
season=pd.read_sql("""SELECT * FROM Season;""",conne)
season
```

```python
matches=pd.read_sql("""SELECT * FROM Match;""",conne)
matches
```

```python
#Matches payed by RCB in 2016
RCB_Mattches=pd.read_sql("""
SELECT Match_Id,Team_2 As Away_Team, Toss_Winner, Match_Winner
FROM Match
WHERE Team_1= (SELECT Team_1 FROM Match WHERE Team_1==2 AND Season_Id==9);""",conne )
print("Matches played by Royal Challengers Bengaluru in 2016 are: ")
RCB_Mattches
```

```python
#Matches won by RCB
RCB_Won_Matches=pd.read_sql("""SELECT * FROM Match
WHERE Match_Winner==2 AND Season_Id=8;""", conne)
print("Matches won by RCB In 2016: ")
RCB_Won_Matches
```

```python
#Who was the match_winner in  a season, teams that played
match_winners=pd.read_sql("""SELECT m.Match_Id,
t1.Team_Name AS Team_1,
t2.Team_Name AS Team_2,
t3.Team_Name AS Match_Winner
FROM Match m
JOIN Team t1 ON m.Team_1=t1.Team_Id
JOIN Team t2 ON m.Team_2=t2.Team_Id
JOIN Team t3 ON m.Match_Winner=t3.Team_Id
WHERE m.Season_Id=8;""",conne)
match_winners
```

```python
#Man of the match compared to whether the team won the match
MOM=pd.read_sql("""
SELECT
 Match_Id, Team_1, Team_2, Match_Winner, Man_Of_The_Match
  FROM Match WHERE Match_Winner=Man_Of_the_Match;""", conne )
MOM
```
